```python
def FrequentWords(Text, k):
    
    freqMap = {}
    n = len(Text)
   
    for i in range(n-k+1):
        Pattern = Text[i:k+i]
        if Pattern not in freqMap:
            freqMap[Pattern] = 1
        else:
            freqMap[Pattern]=freqMap.get(Pattern)+1
      
    sortedFreq = sorted(freqMap.items(), key=lambda x: x[1], reverse=True)
    topKmers = [(kmer, count) for kmer, count in sortedFreq[:3]]
    
    return topKmers
```

```python
def eval(databaselen,seqlen):
    return databaselen*seqlen*((0.25)**seqlen)
```

```python
k = 5
file = open("chr3CAOV3.txt","r")
data = file.readlines()
genome = data[0]

output1 = FrequentWords(genome,k)
output1 
```

```python
for i in range(5,16):
    file = open("chr3CAOV3.txt","r")
    data = file.readlines()
    genome = data[0]

    output1 = FrequentWords(genome,i)
    print("For i = " + str(i) + ",the top three kmers are: "+ "\n")
    for output in output1:
    
        print(output)
    print("E-value: " + str(eval(len(genome),i)))
```
